### 🎯 Objetivo
Aplicar variables, operadores, condicionales, bucles y estructuras de datos en un proyecto sencillo pero completo.

### 💡 Propuesta de mini proyectos:

#### 1. Sistema de gestión de tareas (To-Do List en consola)

- Guardar tareas en una lista o diccionario.

- Poder añadir, marcar como completada, eliminar y mostrar tareas.

- Usar bucles para menús interactivos.

- Usar condicionales para las opciones del usuario.

#### 2. Juego del Adivina el Número

- Generar un número aleatorio.

- Usar bucles while para dar intentos al jugador.

- Condicionales para decir si el número es mayor o menor.

- Contador de intentos.

#### 3. Sistema de calificaciones

- Guardar las notas de varios estudiantes en un diccionario.

- Calcular promedio, nota mayor y menor.

- Mostrar estadísticas usando estructuras de datos.

```python
# 1. Sistema de gestión de tareas (To-Do List en consola)

# Guardar tareas en una lista o diccionario.
# Poder añadir, marcar como completada, eliminar y mostrar tareas.
# Usar bucles para menús interactivos.
# Usar condicionales para las opciones del usuario.

# Creo variables de tareas pendiente en forma de lista
tarea_pendiente = [] # lista vacia de tarea

# Creo Variable de tareas completadas
tarea_completada = []

#Creo Bucle While para gestionar las opciones 
while True:
    print('Bienvenido al sistema de gestion de tareas')
    print('Opcion 0 Tareas Pendiente')
    print('Opcion 1 Agregar Tarea')
    print('Opcion 2 Eliminar Tarea')
    print('Opcion 3 Tareas Completada')
    print('Opcion 4 Mostrar Tarea')
    print('Opcion 5 Salir')

    # manejo de errores de la variable opcion 
    try: 
        opcion = int(input('Seleccione una opcion entre 0 al 5'))
    except ValueError:
        print('❌ Error debe ingresar unicamente numeros enteros entre 0 y 5')
        continue
    if (0 > opcion) or (opcion > 5):
        print('❌Error debe ingresar unicamente valores entre 0 y 5') 
        continue
    
    # Tareas pendiente
    elif opcion == 0:
        if not tarea_pendiente:
            print('No tiene Tarea pendientes')
        else:
            print('Lista de Tareas Pendientes')
            for n,tarea in enumerate(tarea_pendiente):
                print(f'{n} ➡️ {tarea}')

    # Agregar Tarea a la Lista 
    elif opcion == 1:
        tarea_pendiente.append(input('Ingrese la tarea pendiente').lower())
        print(f'Tarea agregada como pendiente: {tarea_pendiente}')
    
    # Eliminar Tarea 
    elif opcion == 2:
        # tarea eliminada
        tarea_eliminada = input('Ingrese tarea a eliminar')
        
        #manejo de Errores
        try: 
            tarea_pendiente.remove(tarea_eliminada)
            print(f'Actualizando lista de tarea despues de la eliminacion {tarea_pendiente}')
        except ValueError:
            print(f'❌ Error La tarea {tarea_eliminada}, no existe en la lista de tarea pendiente')
            continue
        
    # Tarea Completada
    elif opcion == 3:
        # tarea completada
        completa = input('Ingrese la tarea completada')
        try:
            tarea_pendiente.remove(completa)
            print(f'Usted a cumplido con exito la tarea pendiente {completa}👍')
            tarea_completada.append(completa)
        except:
            print(f'❌ Error La tarea {completa}, no existe en la lista de tarea pendiente')
            continue
    
    # mostrar tareas 
    elif opcion == 4:
        
        if not tarea_pendiente:
            print('No Existe Tarea Completada ni Pendiente🔎')
            continue
        # tareas pendientes
        print(f'\nLista de Tareas Pendientes')
        for n1, tarea1 in enumerate(tarea_pendiente):
            print(f'{n1} ➡️ {tarea1}')
        
        # tareas completas
        print(f'\nLista de Tareas Completada')
        for n2, tarea2 in enumerate(tarea_completada):
            print(f'{n2} ➡️ {tarea2}')

    # Cierre del programa
    else:
        print('👋 Hasta Luego, Nos Vemos Pronto')
        break
        
```

```python
# #### 2. Juego del Adivina el Número
#- Generar un número aleatorio
#- Usar bucles while para dar intentos al jugador.
#- Condicionales para decir si el número es mayor o menor.
#- Contador de intentos.

# Importamos el modulo random para generar el numero al azar 
import random 

# creo el intervalo de numeros adivinar 
n_min  = 0
n_max  = 99

# Creo el numero al azar para adivinar
numero = random.randint(n_min,n_max)

# creo la variable intento 
intento = 0

# Bloque While para iniciar el juego

print('Bienvenido al Juego de la adivinanza ')
while True:
    intento +=1 # indica el numero de intento

    # Creo Variable numero_adivinado y manejo de errores.
    try:
        numero_advinado = int(input('⁉️Ingrese un numero del 0 al 99⁉️'))
    except ValueError:
        print('❌❌ Error debe ingresar unicamente numeros enteros')
        continue

    # Manejor de numero negativos y muy altos 
    if (0 > numero_advinado) or (numero_advinado > 99):
        print('🛑Error Ingrese unicamente numeros de 0 al 100')
        continue

    # condiciones de numero mayor y menor
    elif numero_advinado > numero:
        print('El Numero ingresado es muy alto ⬆️. Ingrese uno menor ⬇️')
    
    elif numero_advinado < numero:
        print('El Numero ingresado es muy bajo ⬇️. Ingrese uno mayor ⬆️')
    
    # Condicion de ganadora
    else:
        print(f'Felicidades 🎉🎉🎉🎉🎉\nIntentos realizados {intento}')
        break
    
    
```
