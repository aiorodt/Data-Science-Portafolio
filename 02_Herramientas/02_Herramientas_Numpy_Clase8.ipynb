## üéØ Objetivos

- Entender las reglas de broadcasting.

- Aplicarlas en operaciones entre arreglos de distinta forma.

- Evitar errores de incompatibilidad.

üìñ Teor√≠a

El broadcasting permite que NumPy realice operaciones entre arreglos de tama√±os distintos, extendiendo autom√°ticamente dimensiones cuando es posible.

Reglas:

1. Se comparan las dimensiones desde el final hacia el inicio.

2. Si son iguales o alguna es 1 ‚Üí compatibles.

3. Si son diferentes y ninguna es 1 ‚Üí error.

üíª Ejemplo

```python
import numpy as np

mat = np.array([[1, 2, 3],
                [4, 5, 6]])

columna = np.array([[10],
                    [20]])

print(mat + columna)  # Broadcasting vertical
```

```python
import numpy as np

# Respuesta
# 1. Explica por qu√© np.ones((3,1)) + np.arange(3) funciona.


ab = np.ones((3,1)) + np.arange(3)

print(ab)
print('\nEsta operaci√≥n funciona por el "Broadcasting" de NumPy.')
print('El arreglo (3,1) y el arreglo (3,) son compatibles porque NumPy estira virtualmente el vector fila np.arange(3) para que se sume a cada fila de la matriz columna np.ones((3,1)), resultando en una matriz (3,3).')
```

```python
# Respuesta
# 2. Crea un tensor 2x3x4 y s√∫male un vector de tama√±o 4.
import numpy as np
np.random.seed(15)

tensor2x3x4 = np.random.randint(1,10,[2,3,4]) # Creacion del tensor
print(f'Tensor 2x3x4 (forma: {tensor2x3x4.shape}):\n{tensor2x3x4}')

vector_4= np.arange(1,5)
print(f"\nVector de tama√±o 4 (Forma: {vector_4.shape}):\n{vector_4}")

# Sumo el Vector mas el Tensor

Sumo = tensor2x3x4 + vector_4

print(f"\nResultado de la suma (Forma: {Sumo.shape}):")
print(Sumo)
```

```python
# Respuesta
# 3. Dise√±a un ejemplo donde falle el broadcasting y arr√©glalo.

import numpy as np

# Arreglo 1: Matriz Columna (3 filas, 1 columna)
A = np.array([[1], 
              [2], 
              [3]])  # Forma (3, 1)

# Arreglo 2: Vector Fila de 2 elementos
B = np.array([10, 20])  # Forma (2,)

print(f"Forma de A: {A.shape}")
print(f"Forma de B: {B.shape}")

# Intento de suma que FALLAR√Å
try:
    suma = A + B
    print(suma)
except ValueError as e:
    print("\n--- ERROR DE BROADCASTING ---")
    print(f"La suma falla porque: {e}")
    print("\n[EXPLICACI√ìN DEL FALLO]:")
    print("NumPy intenta comparar las formas de derecha a izquierda:")
    print("1. El tama√±o 1 de A no coincide con el tama√±o 2 de B.")
    print("2. Las dimensiones deben ser iguales o una debe ser 1 (y aqu√≠ son 1 y 2, y luego 3 y nada).")
   print("3. Como la dimensi√≥n (1) y (2) no coinciden y ninguna es 1 o la misma, el Broadcasting falla.")
```

```python
import numpy as np

# Arreglo 1: Matriz Columna de 2x1 (CORREGIDA)
A_corregido = np.array([[10], 
                        [20]])  # Forma (2, 1)

# Arreglo 2: Vector Fila de 2 elementos (B)
B = np.array([1, 2])  # Forma (2,)

print(f"\n--- SOLUCI√ìN (A cambi√≥ a 2x1) ---")
print(f"Forma de A_corregido: {A_corregido.shape}")
print(f"Forma de B: {B.shape}")

# ¬°Ahora la suma funciona!
suma_arreglada = A_corregido + B
print("\nResultado de la suma arreglada (Broadcasting exitoso):")
print(suma_arreglada)
```
