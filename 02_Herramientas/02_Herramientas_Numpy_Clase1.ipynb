## üìñ Teor√≠a r√°pida

NumPy (Numerical Python) es una librer√≠a que permite manejar grandes vol√∫menes de datos num√©ricos de forma eficiente.
Su estructura principal es el ndarray, que representa arreglos n-dimensionales.

Ventajas frente a las listas:

- M√°s r√°pido.

- Menor consumo de memoria.

- Permite operaciones vectorizadas (sin bucles expl√≠citos).

üíª Ejemplo b√°sico

```python
import numpy as np

# Crear arreglos
lista = [1, 2, 3, 4, 5]
arr = np.array(lista)

print("Lista normal:", lista)
print("Arreglo NumPy:", arr)

# Crear arreglos con funciones
zeros = np.zeros(5)                 # Creando matrix cero
ones = np.ones((2, 3))              # Creando Matrix de 1 
rango = np.arange(0, 10, 2)         
aleatorio = np.random.rand(3, 3)    #  creadno Matrix Aleatoria

print('\n---------------------------------------------')

print("Zeros:", zeros)
print("\nOnes:\n", ones)
print("\nRango:", rango)
print("\nAleatorio:\n", aleatorio)
```
