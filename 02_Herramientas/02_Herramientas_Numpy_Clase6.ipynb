## üìñ Teor√≠a

El subm√≥dulo numpy.linalg permite resolver problemas de √°lgebra lineal de forma eficiente.

Operaciones principales:

- np.dot(A, B) o A @ B ‚Üí producto matricial

- np.linalg.inv(A) ‚Üí inversa

- np.linalg.det(A) ‚Üí determinante

- np.linalg.eig(A) ‚Üí eigenvalores y eigenvectores

- np.linalg.solve(A, b) ‚Üí sistema lineal Ax=b

üíª Ejemplo

```python
import numpy as np

A = np.array([[3, 1], [2, 4]])
b = np.array([7, 10])

print("Determinante de A:", np.linalg.det(A))
print("\nInversa:\n", np.linalg.inv(A))

x = np.linalg.solve(A, b)
print("\nSoluci√≥n del sistema:", x)

valores, vectores = np.linalg.eig(A)
print("\nEigenvalores:", valores)
print("\nEigenvectores:\n", vectores)
```

## üß© Ejercicios

- Calcula la inversa y el determinante de una matriz 3x3.

- Resuelve el sistema lineal:

2x+3y=84
x‚àíy=2

- Obt√©n los eigenvalores de una matriz 2x2 y analiza su significado.

```python
# Respuesta 
import numpy as np
np.random.seed(10)

# - Calcula la inversa y el determinante de una matriz 3x3.

# Creo la Matriz 3x3

Matriz3x3 = np.random.randint(2,12,[3,3])
print(f'La Matriz 3x3 es:\n {Matriz3x3}')

# Calculo la inversa
inversa = np.linalg.inv(Matriz3x3)
print(f'\nInversa de la Matriz:\n {inversa}')

# Calculo de determinante 
determinate = np.linalg.det(Matriz3x3)
print(f'\nEL Determinante de la Matriz es {determinate:.2f}')
```

```python
import numpy as np
# respuesta 

# - Resuelve el sistema lineal:
#2x+3y=84
# x‚àíy=2

# Paso 1 Convierto en matriz los coeficiente de cada variable

a = np.array([[2,3],[1,-1]]) # Matriz Coeficiente

# Paso 2 Convierto las igualdades en un vector matriz 
b = np.array([84,2])

# Paso # 3 Calculo la solucion a la Ecuacion 
solucion = np.linalg.solve(a,b)

x =  solucion[0]
y = solucion[1]

print(f"\nLa soluci√≥n al sistema \n2x+3y=84\nx‚àíy=2 es:")
print(f"\nx = {x:.2f}")
print(f"y = {y:.2f}")
```

```python
# Respuesta 
# - Obt√©n los eigenvalores de una matriz 2x2 y analiza su significado.
import numpy as np
np.random.seed(30)

matriz2x2 = np.random.randint(10,20,[2,2]) # Creacion de Matriz 2x2

print(f'Matriz de 2x2 \n{matriz2x2}')

# Calculando los eigenvalores 
valores, vectores = np.linalg.eig(matriz2x2)
print('\nEigenvalores:\n', valores)
print('\nEigenvectores:\n',vectores)

print('\nSignificado de los Eigenvalores:')
print('Cada eigenvalor indica la magnitud por la cual su eigenvector asociado es escalado (estirado o encogido) al aplicar la transformaci√≥n de la matriz.')

print('\nAn√°lisis de los Eigenvalores: El primer vector se estira por un factor de 1.47, mientras que el segundo vector se estira por un factor de 30.52.')
```
