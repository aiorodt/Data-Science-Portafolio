## üìñ Teor√≠a

El subm√≥dulo numpy.linalg permite resolver problemas de √°lgebra lineal de forma eficiente.

Operaciones principales:

- np.dot(A, B) o A @ B ‚Üí producto matricial

- np.linalg.inv(A) ‚Üí inversa

- np.linalg.det(A) ‚Üí determinante

- np.linalg.eig(A) ‚Üí eigenvalores y eigenvectores

- np.linalg.solve(A, b) ‚Üí sistema lineal Ax=b

üíª Ejemplo

```python
import numpy as np

A = np.array([[3, 1], [2, 4]])
b = np.array([7, 10])

print("Determinante:", np.linalg.det(A))
print("\nInversa:\n", np.linalg.inv(A))

x = np.linalg.solve(A, b)
print("\nSoluci√≥n del sistema:", x)

valores, vectores = np.linalg.eig(A)
print("\nEigenvalores:", valores)
print("\nEigenvectores:\n", vectores)
```

- Resuelve el sistema lineal:

2x+3y=84x‚àíy=2

- Obt√©n los eigenvalores de una matriz 2x2 y analiza su significado.

```python
# Respuesta 
import numpy as np
np.random.seed(10)

# - Calcula la inversa y el determinante de una matriz 3x3.

# Creo la Matriz 3x3

Matriz3x3 = np.random.randint(2,12,[3,3])
print(f'La Matriz 3x3 es:\n {Matriz3x3}')

# Calculo la inversa
inversa = np.linalg.inv(Matriz3x3)
print(f'\nInversa de la Matriz:\n {inversa}')

# Calculo de determinante 
determinate = np.linalg.det(Matriz3x3)
print(f'\nEL Determinante de la Matriz es {determinate:.2f}')
```
