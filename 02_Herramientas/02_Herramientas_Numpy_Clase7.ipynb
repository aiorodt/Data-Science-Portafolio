##  Teor铆a

El m贸dulo np.random permite crear datos para simulaciones o experimentos estad铆sticos.
Principales funciones

| Funci贸n                   | Descripci贸n                      |
| ------------------------- | -------------------------------- |
| `rand(n)`                 | Uniforme entre 0 y 1             |
| `randint(a, b, size)`     | Enteros entre `a` y `b`          |
| `normal(mu, sigma, size)` | Distribuci贸n normal              |
| `choice(lista, size)`     | Selecci贸n aleatoria de elementos |


```python
import numpy as np

np.random.seed(30)  # Semilla reproducible

uniforme = np.random.rand(5)  # valores entre 0 y 1
enteros = np.random.randint(0, 100, 10) # valores entre 0 y 100
normal = np.random.normal(0, 1, 5) # Valores entre aletorios 

print("Uniforme:", uniforme)
print("Enteros:", enteros)
print("Normal:", normal)
```

## З Ejercicios

Simula 100 tiradas de un dado y calcula la frecuencia de cada cara.

Genera 1000 n煤meros con distribuci贸n normal media 50, desviaci贸n 10, y calcula su media.

Usa np.random.choice para simular 10 elecciones al azar de una lista de nombres.

```python
import numpy as np
np.random.seed(20)
# Simula 100 tiradas de un dado y calcula la frecuencia de cada cara.

# Creando lanzamiento del dado
tirada_dado = np.random.randint(1,7,100)

print(f'Lanzamiento del dado\n{tirada_dado}')

# --- CLCULO DE FRECUENCIA ---
# np.unique() con return_counts=True devuelve:
# 1. 'caras': Un arreglo con los valores 煤nicos (1, 2, 3, 4, 5, 6).
# 2. 'frecuencias': Un arreglo con el conteo de cada valor.

caras, frecuencia =np.unique(tirada_dado, return_counts=True)

print('\nFrecuencia de Cada Cara')
# Iteramos sobre los resultados para mostrar la frecuencia de manera legible
for cara, freq in zip(caras,frecuencia):
    print(f"La cara {cara} apareci贸 {freq} veces.")
```

```python
# respuesta 
# Genera 1000 n煤meros con distribuci贸n normal media 50, desviaci贸n 10, y calcula su media.
import numpy as np
np.random.seed(40)

# Distribucion Normal 
distribucion_normal = np.random.normal(loc=50, scale= 10, size=1000)

# Calculo su media 
media = np.mean(distribucion_normal)

print(f'La media de la distribucion normal es: {media:.2f}')
```
