## ðŸ“– TeorÃ­a

Una mÃ¡scara booleana es un arreglo de valores True o False del mismo tamaÃ±o que otro arreglo.
Se usa para filtrar elementos segÃºn condiciones.

Por ejemplo:

```python
import numpy as np

arr = np.arange(10)  
mascara = arr > 5    # Selecciona los numero mayores que cinco 
print("MÃ¡scara:", mascara)
print("Elementos mayores a 5:", arr[mascara])
```

## ðŸ§© Ejercicios

- Genera un arreglo con nÃºmeros del 0 al 20 y selecciona los mÃºltiplos de 3.

- Crea un arreglo de 15 valores aleatorios entre 1 y 100, filtra los mayores que 50 y menores que 80.

-  Calcula el promedio solo de los valores que cumplan cierta condiciÃ³n.

```python
# Respuesta 
import numpy as np
# - Genera un arreglo con nÃºmeros del 0 al 20 y selecciona los mÃºltiplos de 3.

arreglo = np.arange(0,21)
print(f'Datos Numericos:\n{arreglo}')

# seleccion de valores 
mayores_3 = arreglo % 3 == 0

print(f'\nLos numero de los datos numericos divisible entre 3 son:\n{arreglo[mayores_3]}')
```

```python
# Respuesta 
import numpy as np
np.random.seed(40)
# Crea un arreglo de 15 valores aleatorios entre 1 y 100, filtra los mayores que 50 y menores que 80.

valores_aleatorio = np.random.randint(1,100,15) # creando los valores aleatorio
print(f'La Lista de valores es:\n{valores_aleatorio}') 

# Desarrollo Filtro
filtrado = (valores_aleatorio > 50) & (valores_aleatorio < 80)

print(f'\nLos valores seleccionado son:\n{valores_aleatorio[filtrado]}')

```

```python
# Respuesta 
# -  Calcula el promedio solo de los valores que cumplan cierta condiciÃ³n.
import numpy as np
np.random.seed(50)

valores = np.random.randint(1,50,10)
print(f'Valores obtenido:{valores}')

# Selecciono los valores mayores a 36 y menores a 40
filtrado = (valores > 36) & (valores < 40)

print(f'\nEl promedio de los valores filtrados son:\n{np.mean(valores[filtrado])}')
```
