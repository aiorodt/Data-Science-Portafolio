# Calculo Del Promedio De Una Lista

Se pretende crear una lista con una serie de datos por ejemplo edad y se determinara el promedio de la misma Ejemplo

``` python

lista_edad = [10,20,30,50,60]
promedio = (sum(lista_edad))/len(lista_edad)
```
Luego mostramos el resultado


````python
lista_edad = [10,20,30,50,60]
promedio = (sum(lista_edad))/len(lista_edad)

print(f'El promedio de la edad es {promedio:.2f}')
````
