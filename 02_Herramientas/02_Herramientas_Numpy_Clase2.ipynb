```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Operaciones vectorizadas
print(f'Suma a + b = {a + b}')
print(f'\nMultiplicacion a * b = {a * b}')
print(f'\nPotencia de a ** 2 = {a ** 2}')
print(f'\nSumando por un valor a + 10 = {a + 10}')

# Producto punto y transpuesta
matriz = np.array([[1, 2], [3, 4]])
print(f'\nMatriz\n{matriz}')

print("\nTranspuesta:\n", matriz.T)
print("\nProducto punto:", np.dot(a, b))
```
