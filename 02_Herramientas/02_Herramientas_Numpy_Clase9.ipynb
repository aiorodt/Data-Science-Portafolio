## 游닀 Teor칤a

NumPy trabaja con vistas por defecto: muchas operaciones no duplican datos, solo cambian la forma de ver el mismo bloque de memoria.

```python
import numpy as np

arr = np.arange(10)
view = arr[3:7]
copy = arr[3:7].copy()

view[:] = 99
print(arr)  # Cambia el original
print(copy)  # Independiente
```
