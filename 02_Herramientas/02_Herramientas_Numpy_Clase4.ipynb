## üéØ Objetivos

- Calcular estad√≠sticas simples con NumPy.

- Comprender y aplicar broadcasting para operaciones entre arreglos de distinta forma

üíª Ejemplo

```python
import numpy as np

datos = np.random.randint(1, 100, 10) # radint(elige diez numero al azar 1 al 100)
print("Datos:", datos)

print("\nMedia:", np.mean(datos))
print("\nMediana:", np.median(datos))
print("\nDesviaci√≥n est√°ndar:", np.std(datos))

# Broadcasting
mat = np.array([[1, 2, 3],
                [4, 5, 6]])

vector = np.array([10, 20, 30])

print("\nResultado broadcasting:\n", mat + vector)
```

## üíº Ejercicios de la vida real

1. Simulaci√≥n de calificaciones:

- Genera un arreglo de 30 calificaciones (valores entre 0 y 100).

- Calcula media, m√°ximo, m√≠nimo y cu√°ntos aprobaron (‚â• 60).

2. An√°lisis de temperaturas:

- Crea un arreglo con las temperaturas diarias (7 d√≠as, 4 semanas ‚Üí matriz 4x7).

- Calcula el promedio semanal y el d√≠a m√°s caluroso.

```python
# Respuesta 
# 1. Simulaci√≥n de calificaciones:
# - Genera un arreglo de 30 calificaciones (valores entre 0 y 100).
import numpy as np
np.random.seed(10)

notas = np.random.randint(0,100,30)
print(f'Nota de los estudiantes:\n{notas}')

# - Calcula media, m√°ximo, m√≠nimo y cu√°ntos aprobaron (‚â• 60).

# Calculo la media 
media_notas = np.mean(notas)
print(f'\nLa media de las notas de los estudiantes es: {media_notas:.2f} puntos')

# Calculo la Maxima Nota
maximo_notas = np.max(notas)
print(f'\nLa mayor nota es {maximo_notas} puntos')

# Calculo la minima nota
minimo_notas = np.min(notas)
print(f'\nLa minima nota es {minimo_notas} puntos ')

# calculamos cuanto aprobaron
notas_mayor_60 = notas>=60

aprobaron = len(notas[notas_mayor_60])
print(f'\nLa cantidad de estudiantes que abrobaron es {aprobaron} estudiantes')
```

```python
# Respuesta
import numpy as np
# 2. An√°lisis de temperaturas:
# - Crea un arreglo con las temperaturas diarias (7 d√≠as, 4 semanas ‚Üí matriz 4x7).
np.random.seed(32)

temperatura = np.random.randint(-100,100,[4,7]) #matriz de temperatura
print(f'Tabla de Temperatura:\n {temperatura}') 

# - Calcula el promedio semanal 

temperatura_promedio_semanal = np.mean(temperatura,axis=1) # promedio semanal
# Extraemos los valores de la temperatura de la lista
for i, promedio in enumerate(temperatura_promedio_semanal):
    print(f'\nSemana {i+1}: {promedio:.2f} ¬∞C')


# Dia mas Caluroso

maxima_temperatura = np.max(temperatura)

# Usamos argmax para encontrar la posici√≥n (√≠ndice plano) del m√°ximo
posicion_max = np.argmax(temperatura)

# Usamos unravel_index para convertir ese √≠ndice plano en (fila, columna)
posicion_2d = np.unravel_index(posicion_max, temperatura.shape)

semana = posicion_2d[0] + 1
dia_de_semana = posicion_2d[1] + 1

print(f'\nLa Temperatura m√°s alta es: {maxima_temperatura} ¬∞C')
print(f'Ocurri√≥ en la Semana {semana}, D√≠a {dia_de_semana} (√çndice 2D: {posicion_2d})')
```
